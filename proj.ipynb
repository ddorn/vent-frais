```python
import geopandas as gpd
import matplotlib.pyplot as plt

world = gpd.read_file(
gpd.datasets.get_path("naturalearth_lowres")
)# Read world cities
cities = gpd.read_file(
 gpd.datasets.get_path("naturalearth_cities")
)# Read Graticules 
```

```python
world.crs
```

```python
fig, ax = plt.subplots(figsize=(12,10))
world.plot(ax=ax, color="lightgray")
cities.plot(ax=ax, color="black", markersize=10, marker ="o")
ax.set(xlabel="Longitude(Degrees)",
 ylabel="Latitude(Degrees)",
 title="WGS84 Datum (Degrees)")

plt.show()
```

```python
from pyproj import CRS, Transformer
crs_4326 = CRS.from_epsg(4326)
crs_8857= CRS.from_epsg(8857)

transformer = Transformer.from_crs(crs_4326, crs_8857)
```
